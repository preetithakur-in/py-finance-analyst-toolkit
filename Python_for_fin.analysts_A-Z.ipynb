```python
'''Function'''
def get_profit(revenue, expenses):  # Python function
    profit = revenue - expenses
    return profit

print(get_profit(500000, 420000))  # call it
```

```python
'''Default Parameters'''
def get_margin(revenue, expenses, round_result = True):
    margin = (revenue - expenses) / revenue
    if round_result:
        return round(margin, 4)
    return margin

print(get_margin(5000000, 4200000))
print(get_margin(5000000, 4200000, False)) #full decimal
```

```python
'''Multiple return values'''
def analyze_stock(revenue, expenses):
    profit = revenue - expenses
    margin = profit / revenue
    status = "profitable" if profit > 0 else "loss"
    return profit, margin, status

# unpack results
profit, margin, status = analyze_stock(500000, 4200000)
print(f"profit: ${profit:,}")
print(f"margin: {margin:.0%}")
print(f"status: {status}")
```

```python
''' Practical finance function'''
def compound_growth(principal, rate, years):
    future_value = principal * (1 + rate) ** years
    return round(future_value, 2)

print(compound_growth(10000, 0.08, 10))
print(compound_growth(50000, 0.06, 5))
```

```python
''' portfolio_summary '''
def portfolio_summary(ticker, shares, buy_price, current_price):
    invested = shares * buy_price
    current_value = shares * current_price
    gain_loss = current_value - invested
    return_pct = gain_loss / invested
    return gain_loss, return_pct

gl, ret = portfolio_summary("AAPL", 100, 150, 178)
print(f"Gain/Loss: ${gl:,.2f}")
print(f"return: {ret:.0%}")
```

```python
''' Real finance use - calculate & save results '''
import csv

# Read portfolio, calculate values, save new file
with open("portfolio.csv", "r") as infile:
    reader = csv.DictReader(infile)
    result = []

    for row in reader:
        value = float(row["price"]) * int(row["shares"])
        result.append({
        "ticker": row["ticker"],
        "price": row["price"],
        "shares": row["shares"],
        "total_value": round(value, 2)
        })

# save enriched data
with open("portfolio_valued.csv", "w", newline = "") as outfile:
    fields = ["ticker", "price", "shares", "total_value"]
    writer = csv.DictWriter(outfile, fieldnames = fields)
    writer.writeheader()
    writer.writerows(result)
print("valued portfolio saved!")
```

```python
''' Daily returns - most used in finance '''
prices = np.array([178.0, 245.0, 420.0, 185.0, 98.0])

# ddaily returns
daily_returns = np.diff(prices) / prices[:-1]
print(np.round(daily_returns, 2))

# average daily return
print(f"avg return: {np.mean(daily_returns):.2%}")

# volatility (risk)
print(f"volatiliy: {np.std(daily_returns):.2%}")
```

```python
''' compound return '''
daily_returns = np.array([0.05, -0.019, 0.048, 0.037])

# total return over period
total_return = np.prod(1 + daily_returns) -1
print(f"total return: {total_return:.2%}")

# annual_return
annual_return = (1 + total_return) ** (252 / len(daily_returns)) - 1
print(round(annual_return * 100, 2), "%")
print(f"annual return: {annual_return:.2%}")
#252 = trading days in a year
#len(daily_returns) = number of days in your data (here: 4)
```

```python
''' Histogram - Return Distribution '''
# simulate 252 trading days of returns
np.random.seed(42)
returns = np.random.normal(0.001, 0.02, 252)

plt.figure(figsize=(10, 5))
plt.hist(returns, bins=30, color="blue",
         edgecolor="black", alpha=0.7)
'''hat it does:
Plots a histogram → shows how your returns are distributed
Parameters:
returns → your data (e.g., daily returns)
bins=30 → splits data into 30 groups
👉 more bins = more detail, fewer bins = smoother view
color="blue" → fill color of bars
edgecolor="black" → black border around each bar
alpha=0.7 → transparency (0 = invisible, 1 = solid)
'''
plt.axvline(returns.mean(), color="red",
            linestyle="--", label="Mean")
'''
What it does:
Draws a vertical line at the average return
Parameters:
returns.mean() → x-position (mean value)
color="red" → line color
linestyle="--" → dashed line
label="Mean" → used in legend
'''
plt.title("Daily Return Distribution")
plt.xlabel("Daily return")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
''' Seaborn - Heatmap(correlation Matrix) '''
# Minthly returns for 4 stocks
data = pd.DataFrame({
    "AAPL": [0.05, -0.02, 0.08, 0.03, -0.01, 0.06],
    "MSFT": [0.04, -0.01, 0.07, 0.02, -0.02, 0.05],
    "TSLA": [0.08, -0.05, 0.12, 0.04, -0.04, 0.09],
    "AMZN": [0.03, -0.02, 0.05, 0.01, -0.01, 0.04],
})

corr = data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr,
            annot=True,
            cmap="coolwarm",
            fmt=".2f",
            linewidths=0.5)
'''
1. corr
This is your correlation matrix (usually from df.corr())
Shows relationships between variables

2. annot=True
Displays the actual values inside each cell
Without this → only colors, no numbers

3. cmap="coolwarm"
Sets the color scheme
"coolwarm":
  Blue → negative correlation
  Red → positive correlation
White → near zero

4. fmt=".2f"
Formats numbers to 2 decimal places
Example:
0.8532 → 0.85

5. linewidths=0.5
Adds lines between cells
Makes grid easier to read

What it shows
How strongly variables are related:
+1 → strong positive (move together)
0 → no relation
-1 → strong negative (move opposite)

'''
plt.title("stock correlation matrix")
plt.show()
```
